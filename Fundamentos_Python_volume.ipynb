## 0. Fundamentos básicos de Python

### 0.1 Variables y tipos de datos

```
entero = 10          # int
flotante = 3.14      # float
texto = 'Bosques'    # str
booleano = True      # bool

print('entero:', entero, type(entero))
print('flotante:', flotante, type(flotante))
print('texto:', texto, type(texto))
print('booleano:', booleano, type(booleano))
```

### 0.2 Listas y diccionarios

```
especies = ['Pino', 'Encino', 'Mezclado']
print('Lista de especies:', especies)
print('Primera especie:', especies[0])

arbol = {
    'diameter': 25.3,
    'height': 12.5,
    'species': 'Pino'
}
print('Diccionario árbol:', arbol)
print('Diámetro del árbol:', arbol['diameter'])
```

### 0.3 Condicionales `if`

```
diameter = 30

if diameter < 20:
    categoria = 'Árbol pequeño'
elif diameter < 40:
    categoria = 'Árbol mediano'
else:
    categoria = 'Árbol grande'

print('Diámetro:', diameter, '→', categoria)
```

### 0.4 Bucles `for`

```
diameters = [10, 20, 30, 40]

for d in diameters:
    print('Diámetro:', d)
```

### 0.5 Funciones

```
import math

def volumen_cilindro(diameter, height):
    """Calcula un volumen aproximado asumiendo forma cilíndrica (solo como ejemplo)."""
    volumen = (math.pi/40000) * (diameter ** 2) * height
    return volumen

v_ejemplo = volumen_cilindro(25, 10)
print('Volumen aproximado (ejemplo):', v_ejemplo)
```

## 4. Análisis gráfico

```
import matplotlib.pyplot as plt
```

### 4.1 Histogramas

```
variables = ['diameter', 'height', 'density', 'basal_area', 'volume']

for var in variables:
    plt.figure()
    plt.hist(df[var].dropna(), bins=20)
    plt.title(f'Histograma de {var}')
    plt.xlabel(var)
    plt.ylabel('Frecuencia')
    plt.grid(True)
    plt.show()
```

### 4.2 Diagramas de dispersión

```
plt.figure()
plt.scatter(df['diameter'], df['height'])
plt.title('Diámetro vs Altura')
plt.xlabel('diameter')
plt.ylabel('height')
plt.grid(True)
plt.show()

plt.figure()
plt.scatter(df['basal_area'], df['volume'])
plt.title('Área basal vs Volumen')
plt.xlabel('basal_area')
plt.ylabel('volume')
plt.grid(True)
plt.show()
```

### 4.3 Boxplots

```
variables = ['diameter', 'height', 'density', 'basal_area', 'volume']

plt.figure()
plt.boxplot([df[var].dropna() for var in variables], labels=variables)
plt.title('Boxplots de variables principales')
plt.ylabel('Valor')
plt.grid(True)
plt.show()
```
