```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split, cross_val_score, learning_curve, GridSearchCV
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score, precision_score, recall_score, f1_score
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.neural_network import MLPClassifier

# Sample Forestry Data (REPLACE WITH YOUR ACTUAL DATA)
data = {'Elevation': np.random.rand(1000),
        'Aspect': np.random.rand(1000),
        'Slope': np.random.rand(1000),
        'DistanceToWater': np.random.rand(1000),
        'SoilType': np.random.randint(1, 4, 1000),  # Example soil types
        'ForestType': np.random.randint(0, 3, 1000)}  # 0: Pine, 1: Oak, 2: Mixed
df = pd.DataFrame(data)

# Separate features (X) and target (y)
X = df.drop('ForestType', axis=1)
y = df['ForestType']

# Split data into training, validation, and testing sets (stratified)
X_train_val, X_test, y_train_val, y_test = train_test_split(X, y, test_size=0.5, random_state=42, stratify=y)
X_train, X_val, y_train, y_val = train_test_split(X_train_val, y_train_val, test_size=0.5, random_state=42, stratify=y_train_val)

# Scale numerical features
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_val = scaler.transform(X_val)
X_test = scaler.transform(X_test)

# Define the classification algorithms with potential hyperparameters to tune
classifiers = {
    "Logistic Regression": (LogisticRegression(max_iter=1000), {'C': [0.1, 1, 10], 'penalty': ['l1', 'l2']}),
    "Support Vector Machine": (SVC(), {'C': [0.1, 1, 10], 'kernel': ['rbf', 'linear']}),
    "K-Nearest Neighbors": (KNeighborsClassifier(), {'n_neighbors': [3, 5, 7], 'weights': ['uniform', 'distance']}),
    "Decision Tree": (DecisionTreeClassifier(), {'max_depth': [None, 5, 10], 'min_samples_split': [2, 5, 10]}),
    "Random Forest": (RandomForestClassifier(), {'n_estimators': [50, 100, 200], 'max_depth': [None, 5, 10]}),
    "Gradient Boosting": (GradientBoostingClassifier(), {'n_estimators': [50, 100, 200], 'learning_rate': [0.01, 0.1, 1.0]}),
    "Naive Bayes": (GaussianNB(), {}),  # No hyperparameters to tune usually
    "Neural Network": (MLPClassifier(max_iter=1000), {'hidden_layer_sizes': [(64,), (128,), (64, 64)], 'activation': ['relu', 'tanh']})
}


results = {}
for name, (clf, param_grid) in classifiers.items():
    print(f"Training and evaluating {name}...")

    # Hyperparameter Tuning using GridSearchCV
    grid_search = GridSearchCV(clf, param_grid, cv=5, scoring='accuracy', n_jobs=-1)  # n_jobs for parallel processing
    grid_search.fit(X_train, y_train)

    best_clf = grid_search.best_estimator_
    y_pred = best_clf.predict(X_test)

    # Performance Metrics (on test set)
    accuracy = accuracy_score(y_test, y_pred)
    precision = precision_score(y_test, y_pred, average='weighted')
    recall = recall_score(y_test, y_pred, average='weighted')
    f1 = f1_score(y_test, y_pred, average='weighted')

    results[name] = {
        'accuracy': accuracy,
        'precision': precision,
        'recall': recall,
        'f1': f1,
        'best_model': best_clf,
        'best_params': grid_search.best_params_
    }

    # Confusion Matrix (on test set)
    cm = confusion_matrix(y_test, y_pred)
    plt.figure()
    plt.imshow(cm, interpolation='nearest', cmap=plt.cm.Blues)
    plt.title(f"Confusion Matrix - {name}")
    plt.colorbar()
    tick_marks = np.arange(len(np.unique(y)))
    plt.xticks(tick_marks, np.unique(y))
    plt.yticks(tick_marks, np.unique(y))
    plt.xlabel('Predicted Label')
    plt.ylabel('True Label')
    plt.tight_layout()
    plt.show()

    # Learning Curve (on training data, using best model)
    train_sizes, train_scores, test_scores = learning_curve(
        best_clf, X_train, y_train, train_sizes=np.linspace(.1, 1.0, 5), cv=5, scoring='accuracy', n_jobs=-1)

    train_scores_mean = np.mean(train_scores, axis=1)
    train_scores_std = np.std(train_scores, axis=1)
    test_scores_mean = np.mean(test_scores, axis=1)
    test_scores_std = np.std(test_scores, axis=1)

    plt.figure()
    plt.title(f"Learning Curve - {name}")
    plt.xlabel("Training examples")
    plt.ylabel("Score")
    plt.gca().invert_yaxis()

    plt.fill_between(train_sizes, train_scores_mean - train_scores_std,
                     train_scores_mean + train_scores_std, alpha=0.1,
                     color="r")
    plt.fill_between(train_sizes, test_scores_mean - test_scores_std,
                     test_scores_mean + test_scores_std, alpha=0.1,
                     color="g")
    plt.plot(train_sizes, train_scores_mean, 'o-', color="r", label="Training score")
    plt.plot(train_sizes, test_scores_mean, 'o-', color="g", label="Cross-validation score")

    plt.legend(loc="best")
    plt.grid()
    plt.show()


# Print Results
print("\n--- Model Evaluation Summary ---")
for name, metrics in results.items():
    print(f"\n{name}:")
    for metric, value in metrics.items():
        print(f"  {metric}: {value}")

# Example of using the best model (based on test set performance)
best_model_name = max(results, key=lambda k: results[k]['accuracy']) # Or F1, or any other metric you want
best_model = results[best_model_name]['best_model']
print(f"\nBest Model: {best_model_name}")
print(f"Best Hyperparameters: {results[best_model_name]['best_params']}")

# Make predictions on new data (same as before)
```
